### Завдання 1

Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

```python
numbers = input("Уведіть у рядок три натуральні числа:").split()

print("Уведені до списку три натуральні числа: %s" % numbers)

min_ = numbers[0]
max_ = numbers[0]

for i in numbers:
    if int(i) < int(min_):
        min_ = i

for j in numbers:
    if int(j) > int(max_):
        max_ = j

print("Найменше число з-поміж уведених: %s" % min_)
print("Найбільше число з-поміж уведених: %s" % max_)
```

* _розрахунок мінімального та максимального значення в списку випадкової послідовності_

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,10) for i in range(1000)]
k = list_with_random_numbers

def minmax(l):

    len_k = 0
    min_value_k = l[0]
    max_value_k = l[0]
    i = 0

    while len_k < len(l):
        if min_value_k > l[i]:
            min_value_k = l[i]
        i = i + 1
        len_k = len_k + 1

    print("Мінімальне значення випадкової вибірки в списку: %s" % min_value_k)

    for ii in range(len(l)):
        if max_value_k < l[ii]:
            max_value_k = l[ii]

    print("Максимальне значення випадкової вибірки в списку: %s" % max_value_k)

c = minmax(k)
```
